# Order Priority

```python
df["Order Priority"].value_counts()
```

```python
df["Order Priority"].unique()
```

```python
df["Order Priority"] = df["Order Priority"].replace('Critical ','Critical')
```

# Ship Mode

```python
x = df["Ship Mode"].value_counts().index
y = df["Ship Mode"].value_counts().values
```

# Conclusion :

High Order Priority for High Count Products: Products with a high count tend to have a high order priority, indicating that they are in demand or essential for customers.
Regular Air Delivery Dominance: Most products are delivered by regular air, suggesting that it is the preferred or default shipping method for the majority of orders.
Office Supplies Dominance in Regular Air Deliveries: Office supplies products are more frequently delivered by regular air compared to other categories, indicating a high demand or necessity for quick delivery.
Corporate Customer Base: The analysis suggests that there are more customers in the corporate sector, possibly indicating a focus or strong presence in B2B sales.
Highest Sales in Office Supplies: The highest sales are observed in the office supplies category, indicating its significance and popularity among customers.
Paper Products Lead in Office Supplies Sales: Within the office supplies category, paper products have the highest sales, showcasing their importance and popularity among customers.
Yearly Sales Increase Indicating Business Growth: The analysis indicates a yearly increase in sales, suggesting business growth and positive performance over time.
Highest Profit in Technology, Lowest in Furniture: Technology products yield the highest profit margins, while furniture products yield the lowest profit margins, highlighting the profitability of different product categories.
Highest Margin in Office Supplies, Lowest in Furniture: Office supplies products have the highest profit margins, while furniture products have the lowest profit margins, indicating varying levels of profitability across product categories.

# Summary

In summary, the analysis suggests a strong market presence in the corporate sector, with office supplies being the dominant category in terms of sales and profit margins. There is also a trend of yearly sales growth, indicating overall business expansion and positive performance.
